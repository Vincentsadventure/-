```python
# 使用 apply() 方法和 lambda 函式
# 捷運站名原本都有"捷運-"，如:"捷運-板橋站"，我想要刪掉該欄位裡面所有的"捷運-"，留下"板橋站"
df_mrt['name'] = df['name'].apply(lambda x: x.replace('捷運-', ''))

# 指定某個欄位，並取消dataframe轉換成數值型態
lat = df_mrt.loc[df_mrt['name'] == "象山", ["latitude"]]
lat = lat['latitude'].values[0]
lat
```
