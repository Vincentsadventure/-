```python
# 匯入套件
from bs4 import BeautifulSoup as bs
import requests as req
from urllib import parse
```

```python
# 取得台北市行政區列表
url = "https://zip5.5432.tw/cityzip/%E8%87%BA%E5%8C%97%E5%B8%82" 

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url) 

# 指定回傳資料的編碼
res.encoding = 'utf-8'

# 指定 lxml 作為解析器
soup = bs(res.text, "lxml") 
```

```python
# 取得 a 的文字
# 對網頁上想抓取的東西按右鍵檢查

for a in soup.select('td.zip-area a'):
    print(a.get_text())
```

```python
# 建立 list 來放置區域連結
list_district = []

prefix = 'https://zip5.5432.tw'

# 取得 a 的超連結
for a in soup.select('td.zip-area a'):
    list_district.append(prefix + a['href']) # append方法:把東西放到最後面
print(list_district)
```

```python
# 走訪每一個 a link，並印出網頁內文
list_road = []
for index, link in enumerate(list_district):
    res = req.get(link)
    res.encoding = 'utf-8'
    soup_ = bs(res.text, "lxml") #soup_保留字
    for i in soup_.select('td.zip-road a'):
        list_road.append(i.text)
        print(list_road)
```

```python
import json, os, pprint, time, re
# 沒有放置 txt 檔的資料夾，就建立起來
folderPath = 'taipeicityroad'
if not os.path.exists(folderPath):
    os.makedirs(folderPath)
with open(f"{folderPath}/taipeicityroad.json", "w", encoding="utf-8") as file:
        file.write( json.dumps(list_road, ensure_ascii=False) )
```
